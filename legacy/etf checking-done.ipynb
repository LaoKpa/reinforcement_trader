```python
OUTPUT_DIR = "./output"
from os import listdir
from os.path import isfile, join
fileLists = [f for f in listdir(OUTPUT_DIR) if f[-4:]==".out"]
```

```python
stdAvg_dictList = []
for j, tmpMatrix in enumerate(upperLeftMatrix):
    std = np.nanstd(tmpMatrix[0])
    mean = np.nanmean(tmpMatrix[0])
    i = 0
    while np.isnan(std) or np.isnan(mean):
        i+=1
        std = np.nanstd(tmpMatrix[i])
        mean = np.nanmean(tmpMatrix[i])
    stdAvg_dictList.append({"Name": fileLists[j], "Std": std, "Mean": mean})
    print(i, fileLists[j], std, mean)
```

```python
stdAvg_df = pd.DataFrame(stdAvg_dictList)
stdAvg_df
```
