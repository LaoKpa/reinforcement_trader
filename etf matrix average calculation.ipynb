```python
from os import listdir
from os.path import isfile, join
fileLists = [f for f in listdir(OUTPUT_DIR) if f[-4:]==".out"]
```

```python
len(fileLists)
```

```python
for i, fileName in enumerate(fileLists):
    tmpMatrix = pickle.load(open(os.path.join(OUTPUT_DIR, fileName), "rb"))
    if i == 0:
        result = np.array([tmpMatrix])
    else:
        result = np.vstack((result,tmpMatrix[None]))
    print(fileName)
```

```python
for j in range(result.shape[0]):
    a = result[j]
    for i in range(a.shape[0]):
        a[i][:i]=np.nan
        a[i]=np.roll(a[i], -1*i)
    print(fileLists[j])
```

```python
def naninfmean(x):
    result = []
    for i in range(x.shape[1]):
        x = matrixMean[:,i]
        x = x[~np.isnan(x)]
        x = x[x!=np.inf]
        result.append(x.mean())
    return np.array(result)
```

```python
dayMean = naninfmean(matrixMean)
```
