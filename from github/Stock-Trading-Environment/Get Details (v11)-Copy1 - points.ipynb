```python
DIR = "./output/11"
plt.figure(figsize=(20,10))
for i in range(10):
    try:
        if i == 0:
            ax0 = plt.subplot(2, 5, 1)
        else:
            plt.subplot(2, 5, i+1, sharex=ax0, sharey = ax0)
        record = pickle.load(open(path.join(DIR,"sp500_3dim_10k-Training_detailed-ModelNo-"+str(i)+".out"), "rb"))
        df = pd.DataFrame(record)
        plt.scatter(df['date'], df['action'])
    except Exception as e:
        print(e)
```
