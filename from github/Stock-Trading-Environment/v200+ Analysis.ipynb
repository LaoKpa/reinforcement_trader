```python
import pickle
from os import path
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint
import numpy as np
```

```python
def plotGraph(dir_num, labels, title, file_prefix):
    DIR = "./output/"+str(dir_num)
    axx = plt.figure(figsize=(20,10))
    plt.suptitle(title)
    plt.rcParams['figure.facecolor'] = 'white'
    for i in range(10):
        try:
            if i == 0:
                ax0 = plt.subplot(2, 5, 1)
                plt.xlabel("Date")
                plt.ylabel("Value ($)")
                ax = ax0
            else:
                ax = plt.subplot(2, 5, i+1, sharex=ax0, sharey = ax0)

            record = pickle.load(open(path.join(DIR,file_prefix+str(i)+".out"), "rb"))
            df = pd.DataFrame(record)
            plt.axhline(y=300000, color = "grey", linestyle='--')
            plt.xlim((pd.Timestamp('2015-01-01'),pd.Timestamp('2019-12-31')))
            ax.set_xticks([pd.Timestamp('2015-01-01'), pd.Timestamp('2016-01-01'), 
                           pd.Timestamp('2017-01-01'), pd.Timestamp('2018-01-01'),
                           pd.Timestamp('2019-01-01'), pd.Timestamp('2020-01-01')])
            plt.plot(df['date'], df['net_worth'], label = "RL Net Value")
            plt.plot(df['date'], df['buyNhold_balance'], label = "Benchmark Net Value")
            for j in range(3):
                plt.plot(df['date'], [tmp[j] for tmp in df['inventory']], label = labels[j])
            plt.title("Model "+str(i+1))
        except Exception as e:
            print(e)
    plt.legend()
    plt.show()
```

```python
dfDIR = "./output/203"
record = pickle.load(open(path.join(dfDIR,"BRZ+TW+NASDAQ-Training_detailed-ModelNo-500000-0.out"), "rb"))
df = pd.DataFrame(record)
plt.figure(figsize=(15,10))
labels = ["high-^BVSP", "mid-^TWII", "low-^IXIC"]
for j in range(3):
    first_price = df['actual_price'][0][j]
    plt.plot(df['date'], [tmp[j]/first_price*300000 for tmp in df['actual_price']], label=labels[j])
plt.plot(df['date'], df['buyNhold_balance'], label="Benchmark Net Value")
plt.legend()
# plt.plot(df['buyNhold_balance'])
```
