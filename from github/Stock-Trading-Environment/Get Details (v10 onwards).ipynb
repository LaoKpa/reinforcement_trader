```python
import pickle
from os import path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DIR = "./output/10"
plt.figure(figsize=(20,10))
for i in range(10):
    record = pickle.load(open(path.join(DIR,"sp500_actual_10k-Training_detailed-ModelNo-"+str(i)+".out"), "rb"))
    df = pd.DataFrame(record)
    plt.subplot(2, 5, i+1)
    try:
        plt.scatter(df['date'], df['action'])
    except Exception as e:
        print(e)
```
