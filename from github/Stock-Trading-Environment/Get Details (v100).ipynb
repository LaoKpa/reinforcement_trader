```python
import pickle
from os import path
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
DIR = "./output/100"
record = pickle.load(open(path.join(DIR,"sp500+dax+hk-Training_detailed-ModelNo-100000-1.out"), "rb"))
df = pd.DataFrame(record)
# df
for i in range(4):
    plt.plot([tmp[i] for tmp in df['inventory']])
plt.plot(df['net_worth'])
```

```python
DIR = "./output/15/0"
plt.figure(figsize=(20,10))
plt.suptitle("Transactions of the Models")
for i in range(10):
    try:
        if i == 0:
            ax0 = plt.subplot(2, 5, 1)
            plt.xlabel("Date")
            plt.ylabel("Value ($)")
        else:
            plt.subplot(2, 5, i+1, sharex=ax0, sharey = ax0)
        record = pickle.load(open(path.join(DIR,"sp500_deltaReward_0k-Training_detailed-ModelNo-"+str(i)+".out"), "rb"))
        df = pd.DataFrame(record)
        plt.scatter(df['date'], df['action']*df['actual_price'])
        plt.axhline(y=0, color = "grey", linestyle='--')
        plt.axhline(y=10000, color = "grey", linestyle='--')
        plt.plot(df['date'], df['actual_profit'], color = "green")
        plt.plot(df['date'], df['net_worth'], color = 'orange')
    except Exception as e:
        print(e)
```

```python
plt.xlim(pd.Timestamp(2015, 1, 1), pd.Timestamp(2019, 12, 31))
plt.axhline(y=0, color = "grey", linestyle='--')
# plt.axhline(y=10000, color = "grey", linestyle='--')
plt.plot(df['date'], df['actual_profit']/df['buyNhold_balance']*100, color = "green")
# plt.plot(df['date'], df['net_worth'], color = 'orange')
```
